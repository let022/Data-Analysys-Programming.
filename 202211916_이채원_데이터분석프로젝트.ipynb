```python
import pandas as pd
import numpy as np
import random

df1 = pd.read_csv("/content/경기도월별고용동향.csv", encoding="CP949")
df1.info()
```

```python
import matplotlib.pyplot as plt

years = df1['기준년월']
agriculture = df1['산업별-농림어업']
manufacturing = df1['산업별-제조업']
construction = df1['산업별-건설업']
service = df1['산업별-도소매,음식,숙박업']
public_service = df1['산업별-사업,개인,공공서비스업']
transportation_finance = df1['산업별-전기,운수,통신,금융업']

plt.figure(figsize=(10, 6))
plt.plot(years, agriculture, label='a') #농림
plt.plot(years, manufacturing, label='b') #제조
plt.plot(years, construction, label='c') #건설
plt.plot(years, service, label='d') #도소매
plt.plot(years, public_service, label='e') #사업
plt.plot(years, transportation_finance, label='f') #전기

plt.title('Population trends by industry')
plt.xlabel('base year')
plt.ylabel('the number of people')
plt.legend()

plt.show()
```

```python

population = df1['경기도15세이상인구'] / 100
economic_activity = df1['경기도경제활동인구'] / 100

plt.figure(figsize=(10, 6))

# 경기도15세이상인구 그래프
plt.plot(years, population, label='Gyeonggi-do Province population aged 15 or older(1/100)')

# 경기도경제활동인구 그래프
plt.plot(years, economic_activity, label='Gyeonggi-do economically active population (1/100)')

plt.title('Trends in the population aged 15 and over and the economically active population of Gyeonggi-do (1/100)')
plt.xlabel('base year')
plt.ylabel('the number of people(1/100)')
plt.legend()

plt.show()
```

```python

employment_rate = df1['고용률(%)']
unemployment_rate = df1['경기도실업률(%)']

plt.figure(figsize=(10, 6))

# 고용률 그래프
plt.plot(years, employment_rate, label='employment rate')

# 실업률 그래프
plt.plot(years, unemployment_rate, label='the unemployment rate')

plt.title('Trends in Employment and Unemployment Rate in Gyeonggi-do')
plt.xlabel('base year')
plt.ylabel('percentage (%)')
plt.legend()

plt.show()
```
